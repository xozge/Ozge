## 1. Hava durumu modeli

Bu ödev için, geçmiş sıcaklık veri kümesine yeniden bakacaksınız.

```python
import pandas as pd
weather_df = pd.read_csv("https://djl-lms-assets.s3.eu-central-1.amazonaws.com/datasets/weatherHistory.csv")
weather_df.head()
```

**1. Önceki derste olduğu gibi, hedef değişkeninizin görünür sıcaklık ve sıcaklık arasındaki fark olduğu bir doğrusal regresyon modeli oluşturun. Açıklayıcı değişkenler olarak nem ve rüzgar hızı kullanın. Şimdi, modelinizi OLS kullanarak tahmin edin. R-kare ve ayarlanmış R-kare değerleri nelerdir? Tatminkar olduklarını düşünüyor musunuz? Açıklayın.**

```python
Y2= weather_df["Temperature (C)"]-weather_df["Apparent Temperature (C)"]
```

```python
X2 = weather_df[['Humidity','Wind Speed (km/h)']]
```

```python
import statsmodels.api as sm

X3 = sm.add_constant(X2)
results = sm.OLS(Y2, X3).fit()
results.summary()
```

R^2 ve adj. R^2 değerleri çok düşüktür. Model olarak kabul etmemiz içn yeterli değildir.

Model hedefteki varyansı açıklamada yetersizdir.

Model olarak kabul  edebilmemiz için R^2 en az %70 olmalıdır.

R^2 ' de artış meydana geldi. Bu iyiye işaret ama hala yetersiz. %70 ten çok düşük

## 2. Ev fiyatları modeli

Bu alıştırmada, ev fiyatları modelinizde çalışacaksınız.

```python
house_prices_df = pd.read_csv("https://djl-lms-assets.s3.eu-central-1.amazonaws.com/datasets/house_prices.csv", sep = ";")
house_prices_df.head()
```

**1. Ev fiyatları modelinizi tekrar çalıştırın ve F-testi, R-kare, ayarlanmış R-kare, AIC ve BIC kullanarak modelinizin uygunluğunu değerlendirin.**

```python
Y_hp = house_prices_df['SalePrice']
X_hp= house_prices_df[['OverallQual','YearBuilt', '1stFlrSF', 'GrLivArea']]
```

```python
import statsmodels.api as sm

X1_hp = sm.add_constant(X_hp)
results = sm.OLS(Y_hp, X1_hp).fit()
results.summary()
```

R^2 ve adj R^2 değeri model için iyi seviyelerde sayılır, %70 in üstündedir. 

AIC ve BIC değerleri düşük değil ama başka bir modelle karşılaştırınca daha sağlıklı  sonuç verecektir.

Ayni şekilde F statistic değeri 0 ile sonsuz arasında değer aldığı için başka model ile karşılaştırma yapmak daha sağlıklı olacaktır.



**2. Modelinizin tatmin edici olduğunu düşünüyor musunuz? Açıklayın**

R^2 ve adj R^2 değerleri  %70 in biraz üstündedir. bu yüzden iyi bir model diyebilirz.


**3.  Uygunluğunu iyileştirmek için, bazı değişkenleri ekleyerek veya kaldırarak farklı model özelliklerini deneyin.**

```python
house_prices_df['Over_Gr'] = house_prices_df["OverallQual"] * house_prices_df["GrLivArea"]
```

```python
X_hp1= house_prices_df[['OverallQual','YearBuilt', '1stFlrSF', 'GrLivArea', 'Over_Gr']]
```

```python
import statsmodels.api as sm

X2_hp = sm.add_constant(X_hp1)
results = sm.OLS(Y_hp, X2_hp).fit()
results.summary()
```

R^2 ve adj R^2 değerinde artış gözlenmiştir. 

R^2 değeri 0.757 den 0.773 e çıkmıştır.

AIC ve BIC değerleri bir önceki modele göre düşmüştür. 

Ama F statistic azalma eğilimi göstermiştir. Model karşılaştırması yaptığımızda F statistic değerinin yüksek olması model için daha iyidir.

```python
house_prices_df['Over_1st'] = house_prices_df["OverallQual"] * house_prices_df["1stFlrSF"]
```

```python
X_hp2= house_prices_df[['OverallQual','YearBuilt', '1stFlrSF', 'GrLivArea', 'Over_1st']]
```

```python
import statsmodels.api as sm

X3_hp = sm.add_constant(X_hp2)
results = sm.OLS(Y_hp, X3_hp).fit()
results.summary()
```

**4. Denediğiniz her model için, uygun metrikleri alın ve modellerinizi birbiriyle karşılaştırın. Hangi model en iyisidir ve neden?**

'Over_Gr' eklediğimizde 'Over_1st' eklenmiş haline göre hem R^2 , adj R^2 değerleri hem de F statistic değerleri artmıştır.

Bunun yanı sıra 'Over_Gr'' modelimizde AIC ve BIC değerleri de düşmüştür. Bundan dolayı 'Over_Gr' modeli daha iyidir.
