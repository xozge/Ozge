**(3) Does the preparation courses have an impact on exam performance?**

```python
plt.figure(figsize=(10,20))
baslik_font = {'family': 'arial', 'color': 'darkred','weight': 'bold','size': 13 }
eksen_font  = {'family': 'arial', 'color': 'darkblue','weight': 'bold','size': 10 }


plt.subplot(3,1,1)
sns.barplot(st_per["math score"], st_per["test preparation course"])
plt.title("Math Score & Test Preparation Course", fontdict=baslik_font)
plt.xlabel('Math Score', fontdict=eksen_font )
plt.ylabel('Test Preparation Course', fontdict=eksen_font )


plt.subplot(3,1,2)
sns.barplot(st_per["reading score"], st_per["test preparation course"])
plt.title("Reading Score & Test Preparation Course", fontdict=baslik_font)
plt.xlabel('Reading Score', fontdict=eksen_font )
plt.ylabel('Test Preparation Course', fontdict=eksen_font )


plt.subplot(3,1,3)
sns.barplot(st_per["writing score"], st_per["test preparation course"])
plt.title("Writing Score & test preparation course", fontdict=baslik_font)
plt.xlabel('Writing Score', fontdict=eksen_font )
plt.ylabel('Test Preparation Course', fontdict=eksen_font )


plt.show()
```
